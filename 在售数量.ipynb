```python
#在售数量
from matplotlib.font_manager import FontProperties
import os
import matplotlib.pyplot as plt
import matplotlib
import datetime
font=FontProperties(fname='font/Songti.ttc',size=18)
import pandas as pd
data = pd.Series({
    '2019-07-14':82280,
    '2019-07-19':82640,
    '2019-07-26':83134,
    '2019-08-10':84679,
    '2019-08-16':85349,
    '2019-08-24':86392,
    '2019-08-30':87558,
    '2019-09-07':88657,
    '2019-09-14':89822,
    '2019-09-20':90928,
    '2019-09-27':91855,
    '2019-10-04':92153,
    '2019-10-11':92875,
    '2019-10-18':93392,
    #'2019-10-23':93287,
    '2019-10-25':93577,
    '2019-11-01':93214,
    '2019-11-08':93280,
    '2019-11-15':93081
})
dates = []
for date in data.index:
    dates.append(datetime.datetime.strptime(date, "%Y-%m-%d"))
data.index = dates
matplotlib.rcParams['figure.figsize'] = [10, 6.18]
plt.plot(data.index, data)
plt.xticks(rotation=90)
matplotlib.rc('font', size=18)
plt.grid(True)
title = "北京在售数量变化"
plt.title(title, fontproperties = font)
plt.tight_layout()
plt.savefig(os.path.join('fig', title +'.png'))
```
